```python
import pandas as pd
import numpy as np

# Takes the data in GSE110137_counts.gct,
# deletes the column 'NAME', and adds a title to each axis
frame = pd.read_csv('../data/GSE110137_counts.gct', sep='\t', skiprows=2, index_col=1)
del frame['NAME']
frame.columns.names, frame.index.names = ['Antibiotics'], ['Genes']

# Genes that are unnamed are removed
frame = frame[np.all([frame.index != '-', frame.index != 'not available'], axis=0)]
frame = frame[(frame.T != 0).any()]
frame
```

```python
import scipy.stats as stats

# Forms a pseudo reference that is the geometric mean
# of each gene's expression
pseudo_ref = stats.gmean(counts + 1, axis=1)
pseudo_ref
```

```python
# Gets the ratio of each gene's expression to
# that reference value
ratios = counts / pseudo_ref[:, None]
ratios
```

```python
# Takes the median of the ratios for each condition
medians = np.median(ratios, axis=0)
medians
```

```python
# Normalizes each by dividing by the median
norm_counts = counts / medians[None, :]
norm_counts
```

```python
# Adds 1 to each value and applies a log2 transform
log = (norm_counts + 1).apply(np.log2)
log
```

```python
# Getting the top 1000 most variable genes

top_var = np.var(log, axis=1).sort_values(ascending=False)[:1000].index
top_var
```

```python
import matplotlib

# Sets font as Arial
matplotlib.rcParams['font.sans-serif'] = 'Arial'
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Clusters each condition by euclidean distance. Please see
# https://seaborn.pydata.org/generated/seaborn.clustermap.html
# for more documentation
h = sns.clustermap(log.loc[top_var].T, z_score=1, figsize=(13, 10), col_cluster=False, cmap='RdBu_r', yticklabels=True, xticklabels=False,
                   cbar_kws={'pad': 0.19, 'shrink': 0.7, 'label': 'Z-score'}, center=0, cbar_pos=None, cbar=True)

# Removes all tick lines in the figure
for i in h.fig.axes:
    i.tick_params(size=0)

# Saves figure to as clustermap_test.png
plt.savefig('../figures/clustermap_test.png', dpi=300)
plt.show()
```
